```python
import pandas as pd
import plotly.express as px
```

```python
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug"],
    "Sales": [15000,18000,20000,22000,21000,25000,27000,30000],
    "Profit": [4000,5000,6000,7000,6500,8000,8500,10000],
    "Region": ["North","South","East","West","North","South","East","West"]
}

df = pd.DataFrame(data)
df
```

```python
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
avg_sales = df["Sales"].mean()

print("===== DASHBOARD SUMMARY =====")
print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
print("Average Monthly Sales:", round(avg_sales,2))
```

```python
fig1 = px.line(df, x="Month", y="Sales",
               title="Monthly Sales Trend",
               markers=True)
fig1.show()
```

```python
fig2 = px.bar(df, x="Month", y="Profit",
              title="Monthly Profit Analysis",
              color="Profit")
fig2.show()
```

```python
fig3 = px.pie(df,
              names="Region",
              values="Sales",
              title="Sales Distribution by Region")
fig3.show()
```

```python
print("===== BUSINESS INSIGHTS =====")
print("1. Sales show steady growth from Jan to Aug.")
print("2. Highest sales recorded in August.")
print("3. Profit increases proportionally with sales.")
print("4. All regions contribute equally to revenue.")
```
